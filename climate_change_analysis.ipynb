#A function to load dataset file and create two different dataframes

```python
import pandas as pd

def read_worldbank_file(filename):
    # Read the Worldbank format file using pandas
    df_years = pd.read_csv(filename, skiprows=4)

    df_years = df_years.fillna(0)
    df_years=df_years.drop([ 'Country Code', 'Indicator Code'], axis=1)

    df_countries = df_years.transpose()
    return df_years, df_countries
```

```python
df_years, df_countries = read_worldbank_file('/content/drive/MyDrive/Colab Notebooks/statistics/API_19_DS2_en_csv_v2_5346672.csv')
```

# function to get 4 factors for 10 different countries

```python
def getfactors(data, indecitor_name):
   # list of selected countries
    country = ['Finland','Poland','United Kingdom','Pakistan','India','New Zealand','Turkiye','South Africa','China','United States']
    df_data = data[data['Indicator Name'] == indecitor_name]

    result = df_data[df_data['Country Name'].isin(country)]
    result=result.drop(['Indicator Name'], axis=1)
    # getting data for last 50 years for 10 year time interval
    result=result.loc[:, ['Country Name','1980','1990','2000','2010', '2020']]  
    return result
```

```python
ub_population = getfactors(df_years,'Urban population')
elec_power = getfactors(df_years,'Electric power consumption (kWh per capita)')
energy = getfactors(df_years,'Energy use (kg of oil equivalent per capita)')
co2 = getfactors(df_years,'CO2 emissions (kg per PPP $ of GDP)')
```

# function to draw bar graph for each selected factors 

```python
def plot_bar_graph(data, title, filename):
    import matplotlib.pyplot as plt
    import pandas as pd 
    data.plot.bar(x="Country Name", title=title, figsize=(20, 10));
    plt.savefig(filename,bbox_inches='tight')
    plt.show(block=True);
    return 0
```

```python
plot_bar_graph(ub_population,'Urban population in last 50 year','Urban_population.jpg')
```

```python
plot_bar_graph(elec_power,'Electric power consumption (kWh per capita)','elec_power.jpg')
```

```python
plot_bar_graph(energy,'Energy use (kg of oil equivalent per capita)','energy.jpg')
```

```python
plot_bar_graph(co2,'CO2 emissions (kg per PPP $ of GDP)','co2.jpg')
```

# this section create a heat map for all selected factors for 10 countries

```python
def heat_map(data):
  # list of selected countries 
    country_hm = ['Finland','Poland','United Kingdom','Pakistan','India','New Zealand','Turkiye','South Africa','China','United States']
    result_hm = data[data['Country Name'].isin(country_hm)]
    result_hm=result_hm.loc[:, ['Country Name','Indicator Name','2010']]
     # list of selected indecitors
    selected_values = ['Urban population','Electric power consumption (kWh per capita)', 'Energy use (kg of oil equivalent per capita)','CO2 emissions (kg per PPP $ of GDP)']  # the specific values you want to select
    result_hm = result_hm[result_hm['Indicator Name'].isin(selected_values)]
    import seaborn as sns
    # creating povit table for heat map
    temp_df = result_hm.pivot(index='Country Name', columns='Indicator Name', values='2010')
    # Create a heatmap
    sns.heatmap(temp_df, annot=True,fmt=".1f")
    plt.savefig('heatmap.jpg')
    # Show the plot
    plt.show()  
```

```python
heat_map(df_years)
```
